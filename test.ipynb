```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
from io import BytesIO

# 1. Setup - Same headers as before to look like a browser
url = "https://www.cepea.org.br/br/indicador/soja.aspx"
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36'
}

# 2. Fetch the Main Page
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.text, 'html.parser')

target_link = soup.find('a', href=lambda href: href and "id=12" in href)


if target_link and 'href' in target_link.attrs:
    file_url = target_link['href']
    
    # Handle relative URLs (if the link starts with '/')
    if not file_url.startswith('http'):
        base_url = "https://www.cepea.org.br"
        file_url = base_url + file_url
        
    print(f"Found file URL: {file_url}")

    # 4. Download the Excel File
    file_response = requests.get(file_url, headers=headers)
    
    if file_response.status_code == 200:
        
        filename = "cepea_soy_historical_data.xls"
        with open(filename, "wb") as f:
            f.write(file_response.content) # Use .content (bytes), NOT .text
        print(f"File successfully saved as '{filename}'")
    
    else:
        print(f"Error downloading file. Status Code: {file_response.status_code}")

else:
    print("Could not find the download link.")
```
